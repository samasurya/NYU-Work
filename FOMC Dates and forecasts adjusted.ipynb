```python
import pandas as pd
%pylab inline
#importing required packages
```

```python
df1  = pd.read_excel(r'C:\Users\samas\Downloads\ECGDUS4Sai.xlsx', sheet_name= 'BN_SURVEY_AVERAGE') 
```

```python
df1['cyr'] = df1['Date'].dt.year
df1['cqtr'] =  pd.to_datetime(df1.Date).dt.quarter
df1['cpq'] = df1['cyr'].apply(str) + df1['cqtr'].apply(str)
```

```python
df2 = pd.read_excel(r'C:\Users\samas\Downloads\fomc_dates.xlsx') #loading the datasets, replace the path
```

```python
df1['str_date'] = df1['Date'].astype(str)
df2['str_date'] = df2['date'].astype(str)   #converting the dates to strings to iterate
before = [] 
after = [] #an empty list to collect the values
for idx, row in df1.iterrows():  #iterating through the rows of the forecast dataset
    x = 0        #indicator variable
    for i in df2['str_date']:  #looping through dates in the fomc dataset
        if i in row['str_date']:  #checking if the fomc date exists in the forecast dataset
            prev = df1['cpq'][idx-1]
            prev = int(prev)
            a = df1.loc[idx - 1, prev:]  # if yes, locating the previous day forecasts row and getting all the values in that row
            a = a.dropna()
            a = list(a.values)
            #a = a[1:]
            d = a[-1]
            a = a[:-1]
            a = a[:len(a)-3] #dropping all the NaN values and formatting the values as required
            a = [i] + [d] + a #adding the fomc date and the previous day date to the list of values
            prev1 = df1['cpq'][idx+1]
            prev1 = int(prev1)
            b = df1.loc[idx + 1, prev1:]
            b = b.dropna()
            b = list(b.values)
            #b= b[1:]
            e = b[-1]
            b = b[:-1]
            b = b[: len(b) - 3]
            b = [i] + [e] + b
            before.append(a)  #appending the list to the empty list
            after.append(b)
            x = 1   #indicator 1, if date exists
            break
        if x == 0:
            pass  #if fomc date doesnt exist in forecasts, continuing with next date

```

```python
dataset = pd.DataFrame(before)  
dataset1 = pd.DataFrame(after) #creating two dataframes for before and after forecasts
```

```python
dataset1.drop(columns=0, inplace= True) #dropping the repeated column
dataset_final = pd.concat([dataset, dataset1], ignore_index= True,axis = 1) #concating the two datasets
dataset_final.rename(columns= {0:'fomc_date', 1 : 'prev_day_date', 2 : 'gdp_left0', 3:'gdp_left1', 4:'gdp_left2', 5:'gdp_left3', 6:'gdp_left4', 7: 'gdp_left5', 8: 'gdp_left6', 9:'gdp_left7', 10: 'gdp_left8', 11:'next_date', 12:'gdp_right0', 13:'gdp_right1', 14: 'gdp_right2', 15: 'gdp_right3', 16:'gdp_right4', 17:'gdp_right5', 18:'gdp_right6', 19:'gdp_right7', 20:'gdp_right8'}, inplace = True)#renaming the columns
dataset_final.to_csv(r'C:\Users\samas\Downloads\forecasts_avg_gdp_adj.csv') #saving the dataset as a csv file, replace the path
```
