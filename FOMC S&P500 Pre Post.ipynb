```python
import pandas as pd
```

```python
df = pd.read_excel(r'C:\Users\samas\Downloads\FOMC_windows.xls')
```

```python
df
```

```python
mask = (df.time_since_FOMC < 0)
```

```python
df1 = df.loc[mask]
```

```python
df2 = df.loc[~mask]
```

```python
post_fomc = df2.loc[df2.groupby(['date'])['time_since_FOMC'].idxmax()]
```

```python
pre_fomc = df1.loc[df1.groupby(['date'])['time_since_FOMC'].idxmin()]
```

```python
post_fomc = post_fomc[['date', 'time_since_FOMC', 'open', 'high', 'low', 'close' ,'fomc_type', 'Flag']]
```

```python
pre_fomc = pre_fomc[['date', 'time_since_FOMC', 'open', 'high', 'low', 'close' ,'fomc_type', 'Flag']]
```

```python
datasets = [pre_fomc, post_fomc]
sheet_names = ['Pre-FOMC', 'Post-FOMC']
name = 'pre_post_fomc.xlsx'
with pd.ExcelWriter(name) as writer:
    for data, sheet in zip(datasets, sheet_names):
        df = pd.DataFrame(data)
        df.to_excel(writer, sheet_name=sheet)
```
