```python
df1 = pd.read_csv(r"C:\Users\samas\OneDrive\Desktop\BankDirectors.csv")
```

```python
m = (df1.post == 0)
```

```python
df1_pre = df1.loc[m]
```

```python
m1 = (df1.post == 1)
```

```python
df1_post = df1.loc[m1]
```

```python
df2 = pd.read_csv(r"C:\Users\samas\OneDrive\Desktop\FirmDirectors.csv")
```

```python
n = (df2.post == 0)
```

```python
df2_pre = df2.loc[n]
```

```python
n1 = (df2.post == 1)
```

```python
df2_post = df2.loc[n1]
```

```python
def common_in_pre(creditorCode, firmCode):
    a = df1_pre.loc[df1_pre["creditorCode"] == creditorCode ,'corpDirNumDIN'].to_list()
    b = df2_pre.loc[df2_pre['firmCode'] == firmCode , 'corpDirNumDIN'].to_list()
    a_clean = [x for x in a if str(x) != 'nan']
    b_clean = [x for x in b if str(x) != 'nan']
    a1 = df1_pre.loc[df1_pre["creditorCode"] == creditorCode ,'directorName'].to_list()
    b1 = df2_pre.loc[df2_pre['firmCode'] == firmCode , 'directorName'].to_list()
    if len(a1) == 0:
        return -1
    elif len(b1) == 0:
        return -1
    elif len(set(a_clean).intersection(b_clean)) > 0:
        return 1
    elif len(set(a1).intersection(b1)) > 0:
        return 1
    else:
        return 0
    
```

```python
for x in df.index:
    dummy = common_in_pre(df.creditorCode[x], df.firmCode[x])
    df.loc[[x],'common_in_pre'] = dummy
    
```

```python
def common_in_post(creditorCode, firmCode):
    a = df1_post.loc[df1_post["creditorCode"] == creditorCode ,'corpDirNumDIN'].to_list()
    b = df2_post.loc[df2_post['firmCode'] == firmCode , 'corpDirNumDIN'].to_list()
    a_clean = [x for x in a if str(x) != 'nan']
    b_clean = [x for x in b if str(x) != 'nan']
    a1 = df1_post.loc[df1_post["creditorCode"] == creditorCode ,'directorName'].to_list()
    b1 = df2_post.loc[df2_post['firmCode'] == firmCode , 'directorName'].to_list()
    if len(a1) == 0:
        return -1
    elif len(b1) == 0:
        return -1
    elif len(set(a_clean).intersection(b_clean)) > 0:
        return 1
    elif len(set(a1).intersection(b1)) > 0:
        return 1
    else:
        return 0
    
```

```python
for x in df.index:
    dummy = common_in_post(df.creditorCode[x], df.firmCode[x])
    df.loc[[x],'common_in_post'] = dummy
```
