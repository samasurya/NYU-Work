## Questions 2,3,4,5

```python
#Calculating Excess returns for each quarter
all_rets[['VBTLX', 'VWILX', 'VGPMX', 'VTSAX', 'VTRIX', 'VGSLX']] = all_rets[['VBTLX', 'VWILX', 'VGPMX', 'VTSAX', 'VTRIX', 'VGSLX']] - all_rets.VMRXX.values[:, None]
```

```python
all_rets.mean()
```

```python
mean_excess = np.array(all_rets.mean())
mean_excess = mean_excess[0:6]
```

```python
mean_excess
```

```python
all_rets # excess returs
all_rets = all_rets - all_rets.mean() # demeaned excess returns
all_rets.drop(columns = 'VMRXX', inplace = True)
```

```python
all_rets #Dataset with demeaned excess returns and no riskfree assets.
```

## Solution to Q2

```python
# Covariance matrix of six excess returns series
cv_excess_demeaned_rets = all_rets.cov()
cv_excess_demeaned_rets
```

## Solution to Q3

```python
# Correlation Matrix of excess retursn series for six funds
corr_excess_demeaned_rets = all_rets.corr()
corr_excess_demeaned_rets
```

## Solution to Q4

```python
# Computing Inverse Covariance matrix
cv_inv = np.linalg.inv(cv_excess_demeaned_rets)
cv_inv
```

## Solution to Q5

```python
Sharpe_weights = mean_excess.dot(cv_inv) # computing the portfolio optimal weights
```

```python
Sharpe_weights
```

```python
normalized_weights = []
for i in Sharpe_weights:
    i = i/sum(Sharpe_weights)
    normalized_weights.append(i)
```

## Optimal Sharpe portfolio weights

```python
normalized_weights
```

## Calculating Portfolio Variance

```python
portfolio_var = np.array(normalized_weights).dot(cv_excess_demeaned_rets).dot(np.array(normalized_weights).T)
```

## Portfolio Standard Deviation

```python
portfolio_sd = np.sqrt(portfolio_var)
portfolio_sd
```

## Portfolio Expected returns based on mean excess historical returns

```python
portfolio_excess_rets = np.array(mean_excess).dot(np.array(normalized_weights))
```

## Computing Portfolio Sharpe Ratio

```python
portfolio_sharpe = portfolio_excess_rets/ portfolio_sd
portfolio_sharpe 
```

## Solution to Q6

```python
adviser_weights = [0.3, 0.15, 0.1, 0.2, 0.15, 0.1 ]
```

```python
# Calculating Standard Deviation of Adviser Portfolio
adviser_sd = np.sqrt(np.array(adviser_weights).dot(cv_excess_demeaned_rets).dot(np.array(adviser_weights).T))
```

```python
# Calculating Adviser portfolio Variance
adviser_var = np.array(adviser_weights).dot(cv_excess_demeaned_rets).dot(np.array(adviser_weights).T)
```

```python
# Getting the expected returns of six funds according to Advisers Expected return from his/ her portfolio.
adv_exp_rets = (0.06/ adviser_var)*(np.array(adviser_weights)).dot(cv_excess_demeaned_rets)
```

```python
adv_exp_rets
```

```python
mean_excess
```

```python
#Advisers Sharpe Ratio
adviser_sharpe = adv_exp_rets.dot(adviser_weights)/ adviser_sd
adviser_sharpe
```

From the results above we can say that the expected returns of the six funds calculated by the adviser are not that reasonable, since based on historical performance of the funds, the returns should be close to mean excess returns calculated . But the returns used by adviser are quite dfferent based on his own assumptions which we do not know. And the shape ratio of the advisers portfolio is lower than the one computed in previous question. This says that the adviser is getting low return for more risk exposure compared to our portfolio.
