```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\samas\Downloads\2021 Assignment 3 data (1).csv")
```

```python
df.plot(x= 'Maturity', y = 'YTM_ annual')
```

```python
bootstrap = []
for index, row in df.iterrows():
    currentprice = row['Price']
    
    b = []
    for k in range(0, index - 1):
        num1 = df['coupon '][k]
        num2 = (1+ bootstrap[k])**(k+1)
        b.append(num1/num2)
    currentprice = currentprice - sum(b)
    bootstrap.append((((row['coupon '] + row['par'])/currentprice)**(1/(index+1))) - 1)
```

```python
df['bootstrap_zero_coupon'] = bootstrap
```

```python
df['bootstrap_zero_coupon'] = df['bootstrap_zero_coupon'] * 100
```
