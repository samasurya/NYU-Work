```python
df1['nyr'] = df1['Date'].dt.year + 1
df1['cqtr'] =  pd.to_datetime(df1.Date).dt.quarter
df1['cpq'] = df1['nyr'].apply(str) + df1['cqtr'].apply(str)
```

```python
df1['str_date'] = df1['Date'].astype(str)
```

```python
df1
```

```python
one_year_ahead = []
for idx, row in df.iterrows():
    ld = pd.date_range(end = row['str_date'] , periods= 15, closed = 'left')
    ld = list(ld.sort_values(ascending = False).astype(str))
    for i in ld:
        for idx1, row1 in df1.iterrows():
            if i in row1['str_date']:
                next_yr_qrt = row1['cpq']
                nxt_yr_qrt_est = row1[int(next_yr_qrt)]
                before_date = row1['str_date']
                spf_deadline = row['str_date']
                new_df = [next_yr_qrt] + [before_date] + [spf_deadline] + [nxt_yr_qrt_est]
                one_year_ahead.append(new_df)
                break
        else:
            continue
        break
```

```python
len(one_year_ahead)
```

```python
new_df = pd.DataFrame(one_year_ahead, columns = ['Next year quarter','Date before deadline','Deadline','Estimated Value'])
```
