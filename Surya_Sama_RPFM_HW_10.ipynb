## Q2

```python
mask = (df.DATE > '1971-07-01')
```

```python
df1 = df.loc[mask]
```

```python
Var_CAD = []
Var_JPY = []
Var_CHF = []
Var_AUD = []
Var_GBP = []
Var_1Yr = []
Var_3Yr = []
Var_5Yr = []
Var_10Yr = []
Var_20Yr =[]
Var_Gold = []
Var_Nasdaq = []
for col in df1.columns[1:]:
    
    for index, row in df1.iterrows():
        b = []
        for k in range(index-99, index-1):
            l = df[col][k]
            b.append(l)
        m = stat.mean(b)
        s = stat.stdev(b)
        var = m-(1.6449*s)
        if col == 'CAD':
            Var_CAD.append(var)
        elif col == 'JPY': 
            Var_JPY.append(var)
        elif col == 'CHF':
            Var_CHF.append(var)
        elif col == 'AUD':
            Var_AUD.append(var)
        elif col == 'GBP':
            Var_GBP.append(var)
        elif col == '1Yr':
            Var_1Yr.append(var)
        elif col == '3Yr':
            Var_3Yr.append(var)
        elif col == '5Yr':
            Var_5Yr.append(var)
        elif col == '10Yr':
            Var_10Yr.append(var)
        elif col == '20Yr':
            Var_20Yr.append(var)
        elif col == 'Gold':
            Var_Gold.append(var)
        else:
            Var_Nasdaq.append(var)
            
```

```python
Var_95_daily = pd.DataFrame(list(zip(Var_CAD,Var_JPY, Var_CHF, Var_AUD, Var_GBP,Var_1Yr,Var_3Yr, Var_5Yr, Var_10Yr, Var_20Yr, Var_Gold, Var_Nasdaq)), columns=['Var_CAD','Var_JPY', 'Var_CHF', 'Var_AUD', 'Var_GBP','Var_1Yr','Var_3Yr', 'Var_5Yr', 'Var_10Yr', 'Var_20Yr', 'Var_Gold', 'Var_Nasdaq'])
```
