## Q3 Solution

```python

plt.figure(figsize=(20,20))
df.plot(x = 'month-year', y = ['S&P500 dividend yield', '10Y coupon rate'])
plt.style.use('fivethirtyeight')
```

## Based  on the time series plot we can say that the s&p500 dividend yield an the 10Y coupon rate are positively correlated. And we can say that as the coupon rate increased the dividend yield decreased. One possible explanation for this is that as the coupon rates rose the cost of borrowing increased and hence the companies retained much of the earnings thus resulting in the fall of dividend yield.

## Q4 Solution

```python
x = df['10Y real return']
y = df['S&P500 real return']
plt.figure(figsize=(10,10))
plt.scatter(x = x, y = y)
m, b = np.polyfit(x,y,1)
plt.plot(x, m*x + b)
m
```

## The positive relationship between the 10Y real return and the S&P500 real return is justified, since on average the risky assets in this case s&p 500 yield more return than the risk free rate in this case 10Y return, due to risk premium, and the positive regression coefficient of 0.31046358168542015 is also justified in this case.

## Q5 Solution

```python
x = df['10Y real return']
y = df['Monthly inflation']
plt.figure(figsize=(10,10))
plt.scatter(x = x, y = y)
m, b = np.polyfit(x,y,1)
plt.plot(x, m*x + b)

m
```

## There is a negative relation between inflation and real return because the real return is calculated as nominal return - inflation. So as inflation increases the real return decreases as the currency depreciates and the purchasing power decreases and the negative regression coefficient of -0.412192577341017 is justified.
