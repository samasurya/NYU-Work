```python
import pandas as pd
%pylab inline
```

```python
df1 = pd.read_excel(r'C:\Users\samas\Downloads\WILL5000INDFC.xls')
df2 = pd.read_excel(r'C:\Users\samas\Downloads\fomc_dates.xlsx', sheet_name= 'Sheet3')
```

```python
df1['observation_date'] = df1['observation_date'].astype(str)
df2['Date'] = df2['Date'].astype(str)
```

```python
df1.dtypes
```

```python
before = []
after = []
for idx, row in df1.iterrows():
    x = 0
    for i in df2['Date']:
        if i in row['observation_date']:
            prev = df1.loc[idx - 1, :]
            aft = df1.loc[idx + 1, :]
            prev = list(prev.values)
            aft = list(aft.values)
            prev = [i] + prev
            before.append(prev)
            after.append(aft)
            x = 1
            break
        if x == 0:
            pass
        
```

```python
dfb = pd.DataFrame(before)
dfa = pd.DataFrame(after)
df = pd.concat([dfb, dfa], ignore_index= True, axis = 1)
```

```python
df.rename(columns={0:'fomc_date',1:'prev_day_date', 2: 'WILL5000INDFC_before',3:'next_day_date',4:'WILL5000INDFC_after' }, inplace = True)
df.to_csv(r'C:\Users\samas\Downloads\WILL5000INDFC_fomc.csv')
```
